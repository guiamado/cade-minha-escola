```python
dados_escolas = pd.read_csv('dadosLimpos.csv', delimiter =';')
df_desempenhos_escolares = dados_escolas
df_desempenhos_escolares
```

```python
df_notas_matematica = df_desempenhos_escolares.loc[df_desempenhos_escolares['cod_disciplina'] == 35]
df_notas_matematica
```

```python
# calculando as medias de matematica

df_notas_matematica = df_desempenhos_escolares.loc[df_desempenhos_escolares['cod_disciplina'] == 35]

df_conta_matematica = df_notas_matematica.groupby(['cod_escola', 'escola'], as_index=False)['nota_bimestral'].count()
df_conta_matematica

df_soma_matematica = df_notas_matematica.groupby(['cod_escola', 'escola'], as_index=False)['nota_bimestral'].sum()
df_soma_matematica

df_media_matematica = df_soma_matematica
for index, row in df_media_matematica.iterrows():
    df_media_matematica.loc[index, 'media'] = df_soma_matematica.loc[index, 'nota_bimestral'] / df_conta_matematica.loc[index, 'nota_bimestral']

df_media_matematica
```

```python
# Calculando a média geral do Distrito federal de Matematica

df_media_geral_Brasilia_matematica = df_media_matematica['media'].sum() / df_media_matematica['media'].count()
df_media_geral_Brasilia_matematica
```
